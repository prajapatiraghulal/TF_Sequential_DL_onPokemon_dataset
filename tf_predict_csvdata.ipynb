````

from pandas import DataFrame, Series
categories =['Type_1','Color','Egg_Group_1','Body_Style']
for i in categories:
  dist_val = df[i].unique()
  for j in dist_val:
    new_series  = Series(1 if x==j else 0 for x in df[i] )
    df[j]=new_series
  df=df.drop(i, axis=1)
````

splitting data into train and test using pandas sample 



````
data_shuffled = df.sample(df.shape[0])
````

````
data_shuffled.head()
````

````
df_train = data_shuffled[0:int(0.7*data_shuffled.shape[0])]
df_test = data_shuffled[int(0.7*data_shuffled.shape[0]):]
````

````
df_train.shape
````

````
df_test.shape
````

````
df_train_label= df_train['isLegendary']
df_test_label = df_test['isLegendary']

#befor fitting dropping label attribute left in df_train which can cause trouble
df_train = df_train.drop('isLegendary',axis=1)
df_test = df_test.drop('isLegendary',axis=1)
````

````
df_test_label
````

````
#data normalization using sklearns preprocessing do manual also 
#df_train = preprocessing.MinMaxScaler().fit_transform(df_train)
#df_test = preprocessing.MinMaxScaler().fit_transform(df_test)
````

````
#manual normalization  used main dataframe to get max and min 
for i in df_train.columns:
  max = df[i].max()
  min= df[i].min()

  for j in df_train[i].index:
    df_train.loc[j,i]= (df_train.loc[j,i]-min)/(max-min)
  
````

````
df_train.head()
````

````
for i in df_test.columns:
  max = df[i].max()
  min= df[i].min()

  for j in df_test[i].index:
    df_test.loc[j,i]= (df_test.loc[j,i]-min)/(max-min)
````

````
df_test.head()
````

````
#all set normalized
````

````
length = df_train.shape[1]

model = keras.Sequential()
model.add(keras.layers.Dense(500, activation='relu',input_shape=[length,]))
model.add(keras.layers.Dense(2, activation='softmax'))

````

````
model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
````

````
#now fit model
model.fit(df_train, df_train_label, epochs=400)
````

````
#evaluate model
loss_val, accuracy_val = model.evaluate(df_test, df_test_label)
````

````
#predict 
prediction = model.predict(df_test)
prediction = prediction.argmax(axis=1)
prediction[0:5]
````
